```python
import networkx as nx
import csv
from itertools import combinations
```

```python
DATA_FOLDER = "../data/"
CSV_ACTORS_PER_PARAGAPH_PER_YEAR = f"{DATA_FOLDER}actors_per_paragraph_per_year.csv"
```

```python
def get_actor_cooccurrences_by_paragraph_for_year(year: str) -> list[tuple[str, str]]:
    actor_pairs = []

    with open(CSV_ACTORS_PER_PARAGAPH_PER_YEAR, mode ='r')as file:
            csvFile = csv.reader(file)
            next(csvFile, None)  # skip the headers
            for lines in csvFile:
                if lines[0] == year:
                    actors_array = lines[2].split(",")
                    actor_set = set()
                    
                    for actor in actors_array:
                         actor_set.add(actor.strip())

                    for actor_tuple in combinations(actor_set, 2):
                        actor_pairs.append(actor_tuple)
    return actor_pairs
```

```python
G = nx.Graph()
G.add_edges_from(get_actor_cooccurrences_by_paragraph_for_year("1999"))
```
