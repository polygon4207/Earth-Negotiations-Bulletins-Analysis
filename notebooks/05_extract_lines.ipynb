```python
import re
import spacy
import os
```

```python
DATA_FOLDER = "../data/"
PDF_FOLDER = f"{DATA_FOLDER}pdf_data/"
TXT_BLOCKS_FOLDER = f"{DATA_FOLDER}txt_blocks/"
TXT_SENTENCES_FOLDER = f"{DATA_FOLDER}txt_sentences/"
SENTENCES_REGEX = r"([A-Z][^\.!?]*[\.!?])"

# Load English tokenizer, tagger, parser and NER
nlp_model = spacy.load("en_core_web_sm")
```

```python
def extract_sentences_block_txt(file_name: str):
    print(f"Extracting sentences from {TXT_BLOCKS_FOLDER}{file_name}.txt...")
    
    with open(f"{TXT_BLOCKS_FOLDER}{file_name}.txt", "r", encoding="utf-8") as file:
        text = file.read()

    text = text.replace("\n", " ")
    text = re.sub(r"\s+", " ", text).strip()
    sentences = re.findall(SENTENCES_REGEX, text)

    with open(f"{TXT_SENTENCES_FOLDER}{file_name}.txt", "w", encoding="utf-8") as output_file:
        for sentence in sentences:
            if len(sentence.strip()) > 40:
                output_file.write(sentence.strip() + "\n")
```

```python
for pdf in os.listdir(PDF_FOLDER):
    file_name = os.path.basename(pdf)
    print(f"Found file: {file_name}")
    if file_name == ".DS_Store":
        continue

    FILE_NAME = file_name.split(".")[0]
    print(f"Processing {FILE_NAME}...")
    extract_sentences_block_txt(FILE_NAME)

    print(f"{FILE_NAME} processed!\n")
    IS_TITLE_WRITTEN = True
```
